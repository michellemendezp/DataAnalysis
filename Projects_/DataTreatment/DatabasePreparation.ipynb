## Imports

```python
import arff
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

## Auxiliary Functions

```python
def load_kdd_dataset(data_path):
    """Reading of the DataSet"""
    with open(data_path, 'r') as train_set:
        dataset = arff.load(train_set)
    attributes = [attr[0] for attr in dataset["attributes"]]
    return pd.DataFrame(dataset["data"], columns=attributes)
```

```python
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

## Reading of the DataSet

```python
df = load_kdd_dataset("KDDTrain+.arff")
```

```python
train_set, val_set, test_set = train_val_test_split(df, stratify='protocol_type')
```

When working with Machine Learning algorithms, dealing with null values in features requires making a choice. You can:
* Delete the rows containing null values.
* Remove the entire attribute (column) that has null values.
* Replace the null values with a specific value, such as zero, the mean, or another relevant value.

## Sklean APIs

Let's do a quick review of how sklearn APIs work:

**Estimators**: Any object that can estimate a parameter.

* You train the estimator using the fit() method, which always takes a dataset as an argument.
* Any other parameter provided to this method is considered a hyperparameter.

**Transformers**: Estimators that can transform datasets (like Imputer).

* Use the transform() method to perform the transformation.
* These take a dataset as an input parameter.

**Predictors**: Estimators that can make predictions.

* The prediction happens through the predict() method.
* They accept a dataset as input and return a dataset with the predictions.
* Predictors also include a score() method to evaluate the prediction results.

## Transforming Categorical Values to numerical

Before we start, let's retrieve the cleaned dataset and separate the labels from the rest of the data. We don't necessarily want to apply the same transformations to both sets.

```python
X_train = train_set.drop("class", axis=1)
y_train = train_set["class"].copy()
```

Machine Learning algorithms generally require numerical data. Since our dataset contains many categorical values, we need to convert them into numerical data.

```python
X_train.info()
```

There are various ways to convert categorical attributes into numerical ones. One of the simplest methods is provided by the factorize function in Pandas, which transforms each category into a sequential number.

```python
protocol_type = X_train['protocol_type']
protocol_type_encoded, categories = protocol_type.factorize()
```

```python
# We show the codification
for i in range(10):
    print(protocol_type.iloc[i], "=", protocol_type_encoded[i])
```

```python
print(categories)
```

### Advanced Transformations with Sklearn

#### Ordinal Encoding

This makes the same codificaction as factorize from Pandas

```python
from sklearn.preprocessing import OrdinalEncoder

protocol_type = X_train[['protocol_type']]

ordinal_encoder = OrdinalEncoder()
protocol_type_encoded = ordinal_encoder.fit_transform(protocol_type)
```

```python
# We show the codification
for i in range(10):
    print(protocol_type["protocol_type"].iloc[i], "=", protocol_type_encoded[i])
```

```python
print(ordinal_encoder.categories_)
```

The issue with this type of encoding is that certain Machine Learning algorithms, which measure the similarity of two points by distance, will consider that 1 is closer to 2 than to 3. However, for these categorical values, this proximity doesn't make sense. That's why other categorization methods, such as One-Hot Encoding, are often used.

#### One-Hot Encoding

This generate a binary matrix for each category of the attribute

```python
#A sparse matrix only stores the positions of the non-zero values to save memory.

from sklearn.preprocessing import OneHotEncoder

protocol_type = X_train[['protocol_type']]

oh_encoder = OneHotEncoder()
protocol_type_oh = oh_encoder.fit_transform(protocol_type)
protocol_type_oh
```

```python
# We convert the sparse to an array
protocol_type_oh.toarray()
```

```python
# We print the encoding 
for i in range(10):
    print(protocol_type["protocol_type"].iloc[i], "=", protocol_type_oh.toarray()[i])
```

```python
print(ordinal_encoder.categories_)
```

When partitioning the dataset or making predictions with new examples, you might encounter new values for certain categories that can cause an error in the transform() function. The OneHotEncoder class offers the handle_unknown parameter, which can either raise an error or ignore the presence of an unknown categorical feature during transformation (the default is to raise an error).

If you set this parameter to "ignore" and an unknown category is encountered during transformation, the resulting encoded columns for this feature will be all zeros. During inverse transformation, an unknown category will be denoted as None.

```python
oh_encoder = OneHotEncoder(handle_unknown='ignore')
```

## Scaling a DataSet

Before we start, let's retrieve the cleaned dataset and separate the labels from the rest of the data. We don't necessarily want to apply the same transformations to both sets.

```python
X_train = train_set.drop("class", axis=1)
y_train = train_set["class"].copy()
```

Machine Learning algorithms don't perform well when the input features have values in very different ranges. To address this, various scaling techniques are used. It's crucial to remember that these scaling mechanisms should not be applied to the labels.

**Normalization**: The attribute values are scaled to fall within a range between 0 and 1.

**Standardization**: The attribute values are scaled to have a similar distribution but are not confined to a specific range.

It's important to apply these transformations only to the training dataset first. Afterward, the same transformations should be applied to the test dataset for evaluation.

```python
from sklearn.preprocessing import RobustScaler

scale_attrs = X_train[['src_bytes', 'dst_bytes']]

robust_scaler = RobustScaler()
X_train_scaled = robust_scaler.fit_transform(scale_attrs)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=['src_bytes', 'dst_bytes'])
```
