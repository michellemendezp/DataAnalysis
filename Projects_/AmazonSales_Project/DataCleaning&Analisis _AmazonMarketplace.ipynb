# The attack begins: Cleaning process 

After the Data Exporation we arrived to the following points to attack:
Categories are in one column we can separate that information for better understanding, there are some variables like 'discounted_price', 
'actual_price', 'discount_percentage' 'rating' and 'raiting count' that need to be transform into numerical variables. 

```python
#Separating Category into categories for better visibility
temp = df['category'].str.split('|', expand=True)
temp.columns = ['Main_Category', 'Sub_Category', 'Category_3', 'Category_4', 'Category_5', 'Category_6', 'Category_7']  # Adjust based on your data
```

```python
df[['Main_Category', 'Sub_Category', 'Category_3', 'Category_4', 'Category_5', 'Category_6', 'Category_7']] = temp
```

```python
df.tail()
```

```python
#In this example, if there are any non-numeric values in the rating column, they will be converted to NaN. This is useful for ensuring clean numerical data when working with potentially messy datasets.

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# Clean and convert the columns to numerical variables and convert discount percentage to percentage. and Fill Nan with 0
df[['discounted_price', 'actual_price', 'discount_percentage','rating','rating_count']] = df[['discounted_price', 'actual_price', 'discount_percentage','rating','rating_count']].replace({'‚Çπ': '', ',': '', '%': ''}, regex=True).astype(float)
df['discount_percentage'] = df['discount_percentage'] / 100
```
